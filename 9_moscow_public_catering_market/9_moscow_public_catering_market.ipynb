### Сетевое распространение по видам объектов общественного питания

Определим для какого вида объекта общественного питания характерно сетевое распространение?

```python
#object_type_chain = df.groupby(['object_type', 'chain']).agg({'id':'count'})

object_type_chain = df.pivot_table(index='object_type', columns='chain', values='id', aggfunc='count').reset_index()
object_type_chain.columns = ['Тип объекта', 'Сетевой', 'Несетевой']
object_type_chain['Доля сетевых'] = object_type_chain['Сетевой'] / (object_type_chain['Сетевой'] + object_type_chain['Несетевой'])
object_type_chain['Доля несетевых'] = object_type_chain['Несетевой'] / (object_type_chain['Сетевой'] + object_type_chain['Несетевой'])
object_type_chain['Всего'] = object_type_chain['Доля сетевых'] + object_type_chain['Доля несетевых']
object_type_chain = object_type_chain.sort_values(by='Доля сетевых', ascending=False).reset_index(drop=True)
print('Таблица сетевого распространение по видам объектов общественного питания')
display(object_type_chain.style.format({'Доля сетевых':'{0:.1%}', 'Доля несетевых':'{0:.1%}', 'Всего':'{0:.1%}'}))
```

### Анализ сетевых заведений на количество посадочных мест

Проанализируем, что характерно для сетевых заведений: много заведений с небольшим числом посадочных мест в каждом или мало заведений с большим количеством посадочных мест?
Для этого:
- определим сетевые заведения,
- для каждой сети посчитаем количество заведений,
- для каждой сети посчитаем среднее количество посадочных мест,
- разделим полученные сети на 4 группы: много/мало заведений, много/мало посадочных мест,
- построим столбцатую диаграмму (barplot), чтобы сравнить полученные группы,
- построим точечную диаграмму (scatterplot), чтобы посмотреть распределение.


```python
# получаем только сетевые заведения
df_chain = df.query('chain == "да"')
df_chain.head()
```

```python
df_chain_group = df_chain.groupby('object_name_chain').agg({'id':'count', 'number':'mean'}).reset_index().sort_values(by='id', ascending=False)
df_chain_group.columns = ['object_name_chain', 'count', 'number']
print('Количество заведений и мест по сетям')
df_chain_group.head()
```

```python
# посчитаем медиану по количеству заведений в сетях
# медиана = 2 , что малова-то для сетей
# при определении границы много/мало заведений в сетях опираться на медиану не получится (очень низкое значение)
df_chain_group['count'].median()
```

```python
# посчитаем среднее по количеству заведений в сетях
# среднее = 9 - достаточно большое,
# большая разница с медианой из-за наличия супер-сетей с большим количеством заведений
round(df_chain_group['count'].mean(), 0)
```

```python
# уточним медиану для сетей, в которых > 1 заведения
# из-за разнообразия названий заведений kfc/кафе kfc 
# и отсутствия критерия для точного объединения заведений в соответствующу сеть,
# часть заведений могли выпасть из своей сети - исключим их из расчета
# возьмем за границу много/мало заведений в сети - это значение
df_chain_group[df_chain_group['count'] > 1]['count'].median()
```

```python
# посмотрим максимальное количество заведений в сетях для bins функции cut()
round(df_chain_group['count'].max(), 0)
```

```python
# посчитаем медиану по количеству посадочных мест в сетевых заведениях
# т.к. медиана более устойчива к выбросам, 
# лучше взять ее для определение границы много/мало мест в сетевых заведениях
round(df_chain_group['number'].median(), 0)
```

```python
# посмотрим максимальное количество мест в сетевых заведениях для bins функции cut()
round(df_chain_group['number'].max(), 0)
```

```python
#df_chain_group['rest_num_group'] = pd.qcut(df_chain_group['count'], q = [0, 0.75, 1], labels = ['Мало заведений', 'Много заведений'])
df_chain_group['count_cat'] = pd.cut(df_chain_group['count'], bins = [0, 4, 200], labels = ['Мало заведений', 'Много заведений'])
print('Таблица категорий много/мало заведений в сетях')
df_chain_group.head()
```

```python
df_chain_group['number_cat'] = pd.cut(df_chain_group['number'], bins = [0, 45, 310], labels = ['Мало мест', 'Много мест'])
df_chain_group['count_number_cat'] = df_chain_group['count_cat'].astype(str) + ' - ' + df_chain_group['number_cat'].astype(str)
print('Таблица категорий много/мало заведений много/мало мест в сетях')
df_chain_group.head()
```

```python
df_chain_grouped = df_chain_group.groupby(['count_cat', 'number_cat']).agg({'object_name_chain':'nunique'}).reset_index()
df_chain_grouped.columns = ['count_cat', 'number_cat', 'count']
print('Сводная таблица категорий много/мало заведений много/мало мест в сетях')
df_chain_grouped
```

```python
# графики
plt.figure(figsize=(18, 6))

ax1 = plt.subplot(1, 2, 1)
sns.barplot(x='count_cat', y='count', hue='number_cat', data=df_chain_grouped, ax=ax1, color='lightblue') 
#ax.set_xlabel('Тип объекта общественного питания')
ax1.set_xlabel('')
ax1.set_ylabel('Количество посадочных мест', size=12)
ax1.set_title('Сравнение сетей \n по количеству заведений/посадочных мест', size=16)
#ax.set_xticklabels(rotation=45)
ax1.set_xticklabels(ax1.get_xticklabels(), rotation = 0, size=12)
sns.despine() 

ax2 = plt.subplot(1, 2, 2)
sns.scatterplot(x = 'count', y = 'number', hue = 'count_number_cat', data = df_chain_group, ax=ax2)

plt.xlabel('Число заведений', size=12)
plt.ylabel('Количество посадочных мест', size=12)
plt.title('Распределение сетей \n по количеству заведений/посадочных мест', size=16)
sns.despine() 
plt.show()
```

Для сетевых заведений характерно:
- Мало заведений и много мест (более 45)
- Мало заведений (до 4) и мало мест (до 45)


Сетевые заведения с большим количеством заведений (больше 4) распространены значительно меньше. Среди них в большей степени встречаются заведения с малым количеством мест и в меньшей степени - с большим количеством посадочных мест.

Кировоградская улица (6577 мест и 110 объектов) и Каширское шоссе (6171 место и 112 объектов) завершают топ-10, как по количеству посадочных мест, так и по количеству заведений.




Таким образом, наиболее перспективными направлениями для открытия оригинального объекта общественного питания с обслуживанием роботами, могут быть:

Вариант 1:
- Предприятие быстрого обслуживания (как средний по распространенности вид общепита, которым рынок еще недостаточно насыщен (13%), но в то же время может быть востребован в дневное/обеденное время за счет автоматизации и быстрого обслуживания;
- 20 посадочных мест;
- Расположение в районах, в которых присутствуют улицы с наиболее низкой плотностью объектов общественного питания - Таганском районе, Хамовниках, Басманном и Тверском районах;
- Перспектива - развитие в сеть, в случае популярности первого заведения (предприятие быстрого обслуживания больше остальных видов заведений представлены сетями).

Вариант 2:
- Ресторан (так же как средний по распространенности вид общепита (15%), который может быть востребован среди обеспеченной публики, которой не важен средний чек, но интересует все необычное, оригинальное, передовое. В этом случае роботы должны быть не только функциональны, но и атмосферны, посещение подобного заведения должно представлять собой атракцион;
- До 100 посадочных мест;
- Расположение в районах, в которых присутствуют улицы с наиболее низкой плотностью объектов общественного питания - Таганском районе, Хамовниках, Басманном и Тверском районах;
- Так же есть перспектива - развитие в небольшую сеть (до 4-х заведений), в случае популярности первого заведения (рестораны занимают 3-е место по представленности сетями).
