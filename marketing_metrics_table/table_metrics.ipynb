## Рассчет метрик для каждой рекламной кампании для каждой недели

### CTR

```python
df['ctr'] = (df['clicks'] / df['shows']) #.map('{:,.2%}'.format)
# df.head().style.set_properties(subset=['ctr'], **{'background-color': 'gold'}) # .format({'ctr': '{:.3f}'}) 
#df.style.format({'ctr': '{:.3f}'})
```

### CPC

```python
df['cpc'] = (df['cost'] / df['clicks'])
```

### CR

```python
df['cr'] = (df['leads'] / df['clicks'])
```

### CR2 (из лида в оплату)

```python
df['cr2'] = (df['payments'] / df['leads'])
```

### CPO

```python
df['cpo'] = (df['cost'] / df['payments']).round().replace([np.inf, -np.inf], np.nan)
```

### Выручка (revenue)

```python
df_info
```

```python
def get_revenue(x):
    '''
    Расчет выручки
    
    выручка = количество платежей * доход с курса
    revenue = df['payments'] * df_info['cost'] 
    payments берем из основной таблицы (df)
    cost берем из таблицы "Информация" (df_info)
    '''
    df_info2 = df_info.copy().set_index('project')
    if x['project'] == 'developer':
        revenue = x['payments'] * df_info2.loc['developer', 'cost'] 
    elif x['project'] == 'data-scientist':
        revenue = x['payments'] * df_info2.loc['data-scientist', 'cost'] 
    elif x['project'] == 'python-developer':
        revenue = x['payments'] * df_info2.loc['python-developer', 'cost'] 
    return revenue
```

```python
df['revenue'] = df.apply(get_revenue, axis=1)
# df.head().style.set_properties(subset=['revenue'], **{'background':'gold'}) 
```

### ROMI

```python
def get_romi(x):
    '''
    Расчет romi
    
    romi = (доход - расход) / расход * 100%
    romi = (df['payments'] * df_info['cost'] - df['cost']) / df['cost'] * 100%
    '''    
    df_info2 = df_info.copy().set_index('project')
    if x['project'] == 'developer':
        romi = (x['payments'] * df_info2.loc['developer', 'cost'] - x['cost']) / x['cost']
    if x['project'] == 'data-scientist':
        romi = (x['payments'] * df_info2.loc['data-scientist', 'cost'] - x['cost']) / x['cost'] 
    if x['project'] == 'python-developer':
        romi = (x['payments'] * df_info2.loc['python-developer', 'cost'] - x['cost']) / x['cost']
    return romi
```

```python
df['romi'] = df.apply(get_romi, axis=1) # .map('{:,.0%}'.format)
```

### ROI

```python
def get_roi(x):
    '''
    Расчет roi
    
    roi = (весь доход - весь расход) / весь расход * 100%
    roi = (df['payments'] * df_info['cost'] * df_info['marginality'] - df['cost']) / df['cost'] * 100%
    '''
    df_info2 = df_info.copy().set_index('project')
    if x['project'] == 'developer':
        roi = (x['payments'] * df_info2.loc['developer', 'cost'] * df_info2.loc['developer', 'marginality'] - x['cost']) / x['cost']
    if x['project'] == 'data-scientist':
        roi = (x['payments'] * df_info2.loc['data-scientist', 'cost'] * df_info2.loc['data-scientist', 'marginality'] - x['cost']) / x['cost']
    if x['project'] == 'python-developer':
        roi = (x['payments'] * df_info2.loc['python-developer', 'cost'] * df_info2.loc['python-developer', 'marginality'] - x['cost']) / x['cost']
    return roi
```

```python
df['roi'] = df.apply(get_roi, axis=1) #.map('{:,.0%}'.format)
```

## Агрегированные метрики  - по каждому проекту (CTR, CPC, CR, CR2, CPO, выручка, ROMI, ROI)

```python
df_project = (
    df.groupby('project', as_index=False)
    .agg({'shows':'sum', 'clicks':'sum', 'cost':'sum', 'leads':'sum', 'payments':'sum'})
    .assign(ctr = lambda x: x['clicks'] / x['shows'],
            cpc = lambda x: x['cost'] / x['clicks'],
            cr = lambda x: x['leads'] / x['clicks'],
            cr2 = lambda x: x['payments'] / x['leads'],
            cpo = lambda x: x['cost'] / x['payments'],
            revenue = lambda x: x.apply(get_revenue, axis=1),
            romi = lambda x: (x['revenue'] - x['cost']) / x['cost'],
            roi = lambda x: x.apply(get_roi, axis=1)
           )
).sort_values(by='project')
df_project = df_project[['project'] + df.columns[6:].to_list()]
# df_project
```

```python
def show_plot(df, x):
    for y in df.columns.to_list()[1:]: # df.columns.to_list()[1:][::-1]        
        fig, ax = plt.subplots(figsize=(10, 4))
        plt.ticklabel_format(style='plain') # убирает le7 и le6 по осям
        ax.bar(df[x], df[y])        
        #ax.set_xticks(df[x])
        #ax.set_xticklabels(df[x], rotation=20, horizontalalignment='right')
        ax.bar_label(ax.containers[0], label_type='edge', padding=3, fmt='%.2f', fontsize=9) # fmt='%.0f' - убирает le7 в подписях столбцов, color='red', rotation=90, fontsize=7, use_scientific=False
        ax.set_title(y)
        sns.despine()
        plt.grid(axis='y')        
        plt.show()
        print(end ='\n\n')
```

```python
print('Метрики по проектам')
show_plot(df_project, 'project')
print('Таблица метрик по проектам')
display(df_project)
```

```python
def get_revenue_campaign(x):
    '''
    Расчет выручки для кампании
    
    выручка = количество платежей * доход с курса
    revenue = df['payments'] * df_info['cost'] 
    payments берем из основной таблицы (df)
    cost берем из таблицы "Информация" (df_info)
    '''
    df_info2 = df_info.copy().set_index('project')
    if '_developer_' in x['campaign']:
        revenue = x['payments'] * df_info2.loc['developer', 'cost'] 
    elif '_data-scientist_' in x['campaign']:
        revenue = x['payments'] * df_info2.loc['data-scientist', 'cost'] 
    elif '_python-developer_' in x['campaign']:
        revenue = x['payments'] * df_info2.loc['python-developer', 'cost'] 
    return revenue
```

```python
# romi = (доход - расход) / расход * 100%
# romi = (df_campaign['revenue'] - df_campaign['cost']) / df_campaign['cost'] * 100%
```

```python
def get_roi_campaign(x):
    '''
    Расчет roi для кампании
    
    roi = (все доходы - все расходы) / все расход * 100%
    roi = (df_campaign['payments'] * df_info['cost'] * df_info['marginality'] - df_campaign['cost']) / df_campaign['cost'] * 100%
    payments берем из основной таблицы (df)
    cost берем из таблицы "Информация" (df_info)
    '''
    df_info2 = df_info.copy().set_index('project')
    if '_developer_' in x['campaign']:
        roi = (x['payments'] * df_info2.loc['developer', 'cost'] * df_info2.loc['developer', 'marginality'] - x['cost']) / x['cost']
    elif '_data-scientist_' in x['campaign']:
        roi = (x['payments'] * df_info2.loc['data-scientist', 'cost'] * df_info2.loc['data-scientist', 'marginality'] - x['cost']) / x['cost'] 
    elif '_python-developer_' in x['campaign']:
        roi = (x['payments'] * df_info2.loc['python-developer', 'cost'] * df_info2.loc['python-developer', 'marginality'] - x['cost']) / x['cost'] 
    return roi
```

```python
def get_relplot(df, col, row, style):
    '''
    График (sns.relplot) динамики показателей по неделям.
    
    df - нужный датафрейм
    col - что пойдет в колонки графика (например, 'project')
    row - что пойдет в строки графика (например, 'medium')
    style - по какому признаку отличать линии на графике (например, по 'source')
    '''
    sns.set_style("whitegrid")
    for i in df.columns[6::]: # df.columns[6::][::-1]
        grid = (sns.relplot(kind='line', 
                     data=df, 
                     x=df['week'], 
                     y=df[i], 
                     col=col, 
                     row=row, 
                     style=style, 
                     height=3, 
                     aspect=1.5, 
                     facet_kws=dict(sharex=False)) # facet_kws=dict(sharey=False) 
        )
        for ax in grid.axes.flat: # или  grid.axes.flatten() - обращаемся к каждому ax
            if i in ('roi', 'romi'):
                ax.axhline(0, color='r', ls='-.', ) # ls='--',
                ax.text(0.5, 0, i, horizontalalignment='left', verticalalignment='bottom', color='red') # ha='center'
        #ax2.set_xlabel('')
        #ax2.set_ylabel('Количество покупателей')
        #ax2.set_xticks(users_months.payment_dt_month.astype('str')) # иначе не заработает ax2.set_xticklabels
        #ax2.set_xticklabels(users_months.payment_dt_month, rotation=0)
        # plt.legend()
        plt.suptitle(i, size=16)    
        plt.tight_layout()
        plt.show()
```
