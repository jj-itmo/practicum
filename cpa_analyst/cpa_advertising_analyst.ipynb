```python
def get_plot_cpa(data, x, y, hue=None, title=None, color=None):
    '''
    Отрисовка графика динамики CPA в зависимости от среза.
    
    data - dataframe, группировка по месяцу и/или срезу (наример, по device/place и т.д.)
    x - str, колонка дат (='month')
    y - float, колонка CPA
    hue - str, колонка среза (=device/place и т.д.)
    title - str, уникальный заголовок графика, отражающий срез
    color - str, цвет графика
    '''
    fig, ax = plt.subplots(figsize=(20, 5))
    sns.barplot(data=data, x=data[x].dt.date, y=data[y], hue=hue, color=color).axhline(380, color = "grey", linestyle = "dashed")
    ax.set_xlabel('')
    ax.set_ylabel('')
    ax.set_title(title, size=14)
    for i in range(len(ax.containers)):
        ax.bar_label(ax.containers[i])
    
    sns.despine()
    plt.show()
```

```python
def get_lineplot(data, x='month', y='cpa', hue=None, title=None, x_lines1=2, y_lines1=2, x_lines2=5, y_lines2=5):
    '''
    Отрисовка линейного графика динамики CPA в зависимости от среза (когда в срезе > 3 значений).
    
    data - dataframe, группировка по месяцу и/или срезу (наример, по device/place и т.д.)
    x - str, колонка дат (='month')
    y - float, колонка CPA  (='cpa')
    hue - str, колонка среза (=device/place и т.д.)
    title - str, уникальный заголовок графика, отражающий срез
    x_lines1 и y_lines1 - int, указывают для какого номера граифка отобразать подписи на линии (=2)
    x_lines2 и y_lines2 - int, указывают для какого номера граифка отобразать подписи на линии (=5)
    '''
    fig, ax = plt.subplots(figsize=(20, 5))
    (sns.lineplot(data=data, x=data[x].dt.date, y=data[y], hue=hue, marker='o')
     .axhline(380, color = "grey", linestyle = "dashed"))
    ax.set_xlabel('')
    ax.set_ylabel('')
    ax.set_title(title, size=14)

    #Add data labels to the lines
    # for line in ax.lines:
    #     x_data = line.get_xdata()
    #     y_data = line.get_ydata()
    #     for i in range(len(x_data)):
    #         ax.annotate(y_data[i], (x_data[i], y_data[i]), textcoords="offset points", xytext=(0, 10), ha='center')

    x_data = ax.lines[x_lines1].get_xdata()
    y_data = ax.lines[y_lines1].get_ydata()
    for i in range(len(x_data)):
        ax.annotate(y_data[i], (x_data[i], y_data[i]), textcoords="offset points", xytext=(0, 10), ha='center')

    x_data = ax.lines[x_lines2].get_xdata()
    y_data = ax.lines[y_lines2].get_ydata()
    for i in range(len(x_data)):
        ax.annotate(y_data[i], (x_data[i], y_data[i]), textcoords="offset points", xytext=(0, 10), ha='center')

    sns.despine()
    plt.tight_layout()
    plt.show()
```

## Анализ рекламных кампаний Сигма

### Динамика CPA

Проверим действительно ли за последние 4 месяца (с июня по сент 2022) существует рост CPA по рекламным кампаниям рекламодателя Сигма. А также рассчитаем дополнительные метрики - ctr, стоимость клика, CR в конверсию.

```python
sigma_cpa = df.query('client == "Сигма"').groupby(['month'], as_index=False).agg({'shows':'sum','clicks':'sum', 'conversions':'sum', 'cost':'sum'})
sigma_cpa['cpa'] = round(sigma_cpa['cost'] / sigma_cpa['conversions'])
sigma_cpa['ctr'] = round(sigma_cpa['clicks'] / sigma_cpa['shows'] * 100, 2)
sigma_cpa['cr'] = round(sigma_cpa['conversions'] / sigma_cpa['clicks'] * 100, 2)
sigma_cpa['cost_per_click'] = round(sigma_cpa['cost'] / sigma_cpa['clicks'], 2)
sigma_cpa = sigma_cpa[['month', 'shows', 'ctr', 'clicks', 'cost_per_click', 'cost', 'cr', 'conversions', 'cpa']]
print()
print('Таблица показателей по месяцам')
sigma_cpa
```

```python
get_plot_cpa(sigma_cpa, 'month', 'cpa', title='Динамика CPA', hue=None, color='lightblue')
```

В последние 4 месяца (с июня по сент 2022) действительно наблюдается рост CPA по рекламным кампаниям рекламодателя Сигма.<br>
Стоит отметить, что целевой CPA не достигался не только в последние 4 месяца, но и в предыдущие периоды.

### Динамика показателей (показы, клики, ctr, cr, количество конверсий, сумма оплат)

Для составления полной картины, посмотрим динамику остальных показателей - показы, клики, ctr, cr, количество конверсий, сумму оплат.

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(18, 7))

# Показы и клики
sns.lineplot(ax=axs[0,0], data=sigma_cpa, x=sigma_cpa['month'], y=sigma_cpa['clicks'], color='lightblue',  marker = 'o')
axs[0, 0].set_xlabel('')
axs[0, 0].set_ylabel('Клики')
axs[0, 0].get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))
ax2 = axs[0,0].twinx()
sns.lineplot(ax=ax2, data=sigma_cpa, x=sigma_cpa['month'], y=sigma_cpa['shows'], color='salmon', marker = 'o')
ax2.set_ylabel('Показы')
ax2.get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))
axs[0, 0].set_title('Показы и клики')
a_patch = mpatches.Patch(color='salmon', label='Показы')
b_patch = mpatches.Patch(color='lightblue', label='Клики')
plt.legend(handles=[a_patch, b_patch], loc='upper left')

# CTR и CR
sns.lineplot(ax=axs[0, 1], data=sigma_cpa, x=sigma_cpa['month'], y=sigma_cpa['cr'], color='royalblue',  marker = 'o')
axs[0, 1].set_xlabel('')
axs[0, 1].set_ylabel('CR')
#axs[0, 1].get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))
ax2 = axs[0, 1].twinx()
sns.lineplot(ax=ax2, data=sigma_cpa, x=sigma_cpa['month'], y=sigma_cpa['ctr'], color='orange', marker = 'o')
ax2.set_ylabel('CTR')
#ax2.set_ylim((0, 3))
axs[0, 1].set_title('CTR и CR')
a_patch = mpatches.Patch(color='orange', label='CTR')
b_patch = mpatches.Patch(color='royalblue', label='CR')
plt.legend(handles=[a_patch, b_patch], loc='upper left')

# Количество конверсий и сумма затрат
sns.lineplot(ax=axs[1, 0], data=sigma_cpa, x=sigma_cpa['month'], y=sigma_cpa['conversions'], color='lightgreen',  marker = 'o')
axs[1, 0].set_xlabel('')
axs[1, 0].set_ylabel('Количество конверсий')
axs[1, 0].get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))
ax2 = axs[1, 0].twinx()
sns.lineplot(ax=ax2, data=sigma_cpa, x=sigma_cpa['month'], y=sigma_cpa['cost'], color='seagreen', marker = 'o')
ax2.set_ylabel('Сумма затрат')
ax2.get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))
#ax2.set_ylim((0, 3))
axs[1, 0].set_title('Количество конверсий и сумма затрат')
a_patch = mpatches.Patch(color='seagreen', label='Сумма затрат')
b_patch = mpatches.Patch(color='lightgreen', label='Количество конверсий')
plt.legend(handles=[a_patch, b_patch], loc='upper left')

# Цена за клик и CPA
sns.lineplot(ax=axs[1, 1], data=sigma_cpa, x=sigma_cpa['month'], y=sigma_cpa['cost_per_click'], color='violet',  marker = 'o')
axs[1, 1].set_xlabel('')
axs[1, 1].set_ylabel('Цена за клик')
axs[1, 1].get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))
ax2 = axs[1, 1].twinx()
sns.lineplot(ax=ax2, data=sigma_cpa, x=sigma_cpa['month'], y=sigma_cpa['cpa'], color='slateblue', marker = 'o')
ax2.set_ylabel('CPA')
#ax2.get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))
#ax2.set_ylim((0, 3))
axs[1, 1].set_title('Цена за клик и CPA')
a_patch = mpatches.Patch(color='slateblue', label='CPA')
b_patch = mpatches.Patch(color='violet', label='Цена за клик')
plt.legend(handles=[a_patch, b_patch], loc='upper left') 

sns.despine(top=True, right=False, left=False, bottom=False)
plt.suptitle('Динамика показателей', size=16)
plt.tight_layout()
plt.show()
```

## Выводы и рекомендации

### Причины растущего CPA

Причиной растущего CPA является категория **"Расчетно-кассовое обслуживание"**, в котором CPA занимает анамально высокий уровень и на сентябрь 2022 составляет **10 095 р.**<br>
CPA был крайне высок не только в последние 4 месяца (июнь-сентябрь 2022), но и в предыдущий доступный для анализа период (мая 2021 по июнь 2022). **Аномально высокий ровень CPA в этой категории оказывает влияние на общий показатель CPA по всем кампаниям рекламодателя.**<br>
Категории характерен **низкий процент (4%) конверсий** и **высокая доля затрат (45%)**. Она является «дорогой» для всех рекламодателей - высокий CPA (>380 р.) наблюдается у всех, при этом у Сигмы, Дельты и Омеги CPA аномально высокий - ~10 000 р. на сентябрь 2022 г..<br>
Исключение категории «Расчетно-кассовое-оборудование» приведет к нормализации общего показателя CPA по кампаниям в целом (до 250-300 р.), существенной экономии бюджета (45%  или 10 млн р./мес) и незначительной потери в конверсиях (4% или 1800 шт./мес).<br>


Также стоит обратить внимание на категорию **"Ипотечное кредитование"**. Не смотря на то, что в данной категории CPA не является аномально высоким, тем не менее он превышает целевое значение в 380 р. и на сентябрь 2022 составляет **475 р.**<br>

В рамках других категорий, устройств, типов площадок/объявлений/запросов/таргетингов, положение и уровень CPA – благоприятные.

### Рекомендации по снижению и стабилизации CPA до целевого значения (380 руб.)

- **Категория "Расчетно-кассовое обслуживание"**<br>
    В связи с тем, что данная категория приносит всего 4% конверсий, а ее бюджет составляет 45% (т.е. затраты несопоставимы с профитом), уровень CPA запредельно высок (10 095 р. на сентябрь 2022 г.) и аномалия касается всех типов устройств, площадок, объявлений, запросов и таргетингов, рекомендуется отключить данную кампанию полностью. Анализ показал, что высокий уровень CPA в данной категории характерен и другим рекламодателям, поэтому выход на целевые 380 р. за счет оптимизации маловероятен.<br>

    В случае невозможности полного отключения и острой необходимости присутствия данной категории в рекламных активностях, рекомендуется провести оптимизацию кампаний/объявлений в данной категории. В первую очередь обратить внимание на:
    - актуальность содержания объявлений, их соответствие ключевым словам, таргетинги (падение CTR);
    - актуальность и соответствие ссылок на посадочные страницы (падение CR);
    - содержание (соответствие запросам) и работоспособность посадочных страниц/сайта (падение CR).


- **Категория "Ипотечное кредитование"**<br>
    В данной категории рекомендуется провести оптимизацию рекламных кампаний с целью снижения уровня CPA до целевых 380 р.<br>
    Прежде всего рекомендуется оптимизировать настройки для объявлений (CPA выше нормы более, чем в 3 раза):
    - Desktop (device) - Networks (place) - text (banner_type) - none (query_type) - Retargeting (targeting_type);
    - Mobile - Networks (place) - text (banner_type) - none (query_type) - Retargeting (targeting_type);
    - Mobile - Search (place) - text (banner_type) - Competitor (query_type) - Phrase (targeting_type).
    
    Затем (CPA выше нормы в 1,5-2 раза):
    - Desktop (device) - Search (place) - text (banner_type) - Brand/Competitor/Other (query_type) - Phrase (targeting_type);
    - Mobile (device) - Search (place) - text (banner_type) - Competitor (query_type) - Autotargeting (targeting_type);
    - Mobile (device) - Search (place) - text (banner_type) - Other (query_type) - Phrase (targeting_type);
    - Mobile (device) - Networks (place) - text (banner_type) - none (query_type) - Phrase (targeting_type).

    При оптимизации, в первую очередь следует обратить внимание на:
    - актуальность содержания объявлений, их соответствие ключевым словам, таргетинги (отсутствие роста кликабельности на фоне роста количества показов, сильные колебания CTR);
    - настройки ставок (резкий рост стоимость клика в июне 2022);
    - актуальность и соответствие ссылок на посадочные страницы (падение CR);   
    - содержание (соответствие запросам) и работоспособность посадочных страниц (падение CR).
